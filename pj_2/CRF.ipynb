```python
import os
import sklearn_crfsuite
from check import *
from NER_dataset import *
from word2features import *
from utils import *


def CRF_check(language):
    train_data, valid_data = get_data_set(language)
    train_features, train_labels = get_features(train_data)

    # 建立并训练CRF模型
    model = sklearn_crfsuite.CRF(
        algorithm='lbfgs',
        c1=0.1,
        c2=0.1,
        max_iterations=1000,
        all_possible_transitions=True
    )
    try:
        model.fit(train_features, train_labels)
    except AttributeError:
        pass

    # 输出
    test_features, test_labels = get_features(valid_data)
    pred = model.predict(test_features)

    my_path = 'my_{language}_CRF_result.txt'
    gold_path = f'./NER/{language}/validation.txt'

    file = open(my_path, "w")
    for idx, words_tags in enumerate(valid_data):
        words, _ = words_tags
        tags = pred[idx]
        for i in range(len(words)):
            file.write(f'{words[i]} {tags[i]}\n')
        file.write('\n')
    file.close()

    check(language, gold_path, my_path)


```

```python
print('---------------- Chinese Result ------------------')
CRF_check('Chinese')
print('---------------- English Result ------------------')
CRF_check('English')
```
